# 讀取檔案

### 讀取CSV

import pandas as pd  
**pd.read_csv()**  

sep: 指定分隔符號（資料要怎麼切割成一格一格） 
預設為 ',' 

header: 設定列名稱  
預設為檔案最上方的一列作為列名稱  
若檔案沒有給列名稱: header = None  

names: 幫沒有列名稱的檔案設定列名稱  
需寫入 header = None  

skiprows: 讀取時要略過最上方幾列的數據  
*若資料原本就有列名稱，列名稱也算一列

 


```python
import pandas as pd
```

```python
csv_url = "https://storage.googleapis.com/ds_data_import/chicago_bulls_1995_1996.csv"
csv_df = pd.read_csv(csv_url)
csv_df.head(3)
```

```python
csv_df = pd.read_csv(csv_url,header = 0)
csv_df.head(3)
```

```python
csv_df = pd.read_csv(csv_url,header = None)
csv_df.head(3)
```

```python
csv_df = pd.read_csv(csv_url,header = None,names=['A','B','C','D','E','F','G'])
csv_df.head(3)
```

```python
csv_df = pd.read_csv(csv_url,header = None,names=['A','B','C','D','E','F','G'],skiprows=2)
csv_df.head(3)
```

### 讀取TXT

import pandas as pd  
**pd.read_table()**

sep: 指定分隔符號（資料要怎麼切割成一格一格） 
預設為 '\t' 

```python
txt_url = "https://storage.googleapis.com/ds_data_import/chicago_bulls_1995_1996.txt"
txt_df = pd.read_table(txt_url, sep=";")
txt_df.head(3)
```

### 讀取Excel

import pandas as pd  
**pd.read_excel()**

sheet_name: 可以設定要讀取檔案中的哪一張工作表  
預設為0

```python
xlsx_url = "https://storage.googleapis.com/ds_data_import/fav_nba_teams.xlsx"
chicsgo_bulls = pd.read_excel(xlsx_url)
chicsgo_bulls.head(3)
```
